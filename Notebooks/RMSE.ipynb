```python
tickers = ['TECHM.NS', 'M&M.NS', 'WIPRO.NS', 'SBIN.NS', 'BAJFINANCE.NS',
    'INFY.NS', 'KOTAKBANK.NS', 'TRENT.NS', 'BAJAJFINSV.NS', 'SHRIRAMFIN.NS',
    'ADANIPORTS.NS', 'TATAMOTORS.NS', 'AXISBANK.NS', 'COALINDIA.NS', 'HCLTECH.NS',
    'BRITANNIA.NS', 'ICICIBANK.NS', 'LT.NS', 'CIPLA.NS', 'NTPC.NS',
    'BHARTIARTL.NS', 'NESTLEIND.NS', 'TATACONSUM.NS', 'HEROMOTOCO.NS', 'APOLLOHOSP.NS',
    'BEL.NS', 'HDFCBANK.NS', 'HINDALCO.NS', 'TCS.NS', 'SBILIFE.NS',
    'POWERGRID.NS', 'ULTRACEMCO.NS', 'ADANIENT.NS', 'HDFCLIFE.NS', 'GRASIM.NS',
    'ITC.NS', 'DRREDDY.NS', 'RELIANCE.NS', 'ONGC.NS', 'JSWSTEEL.NS',
    'EICHERMOT.NS', 'TITAN.NS', 'SUNPHARMA.NS', 'INDUSINDBK.NS', 'BPCL.NS',
    'MARUTI.NS', 'TATASTEEL.NS', 'BAJAJ-AUTO.NS', 'ASIANPAINT.NS', 'HINDUNILVR.NS']
```

```python
rmse_arima = []
rmse_lstm = []
```

```python
folder_path = '/content/drive/MyDrive/Ph.D/Datasets/lstm_arima_compare/'
```

```python
for ticker in tickers:
    file_path = os.path.join(folder_path, f'{ticker}.csv')

    df = pd.read_csv(file_path)

    # Calculate RMSE for ARIMA and LSTM
    rmse_arima_value = np.sqrt(mean_squared_error(df['Actual Price'], df['ARIMA Forecast']))
    rmse_lstm_value = np.sqrt(mean_squared_error(df['Actual Price'], df['LSTM Forecast']))

    # Append RMSE values to respective lists
    rmse_arima.append(rmse_arima_value)
    rmse_lstm.append(rmse_lstm_value)

    print(f'process for {ticker} is complete')
```

```python
    plt.figure(figsize=(10, 6))
    plt.plot(tickers, rmse_arima, label='ARIMA RMSE', marker='o', color='blue')
    plt.plot(tickers, rmse_lstm, label='LSTM RMSE', marker='x', color='green')
    plt.xlabel('Company Ticker')
    plt.ylabel('RMSE')
    plt.title('RMSE Comparison for ARIMA and LSTM Models')
    plt.xticks(rotation=90)  # Rotate ticker labels for better readability
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'/content/drive/MyDrive/Ph.D/Datasets/lstm_arima_compare/RMSE/RMSE.png', format='png', dpi=300)
```

```python
rmse_values = pd.DataFrame({'Ticker': tickers, 'ARIMA RMSE': rmse_arima, 'LSTM RMSE': rmse_lstm})
rmse_values.to_csv('/content/drive/MyDrive/Ph.D/Datasets/lstm_arima_compare/RMSE/rmse_values.csv', index=False)
```
