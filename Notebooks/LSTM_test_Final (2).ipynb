```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
import yfinance as yf
```

```python
start = '2014-01-01'
end = '2023-12-31'
```

```python
test_start = '2024-01-01'
test_end = '2024-03-01'
```

```python
#give the number of days to forecast
forecast_days = 20
past_date_to_use = 60
```

```python
# ticker = ['ADANIPORTS.NS']
```

```python
ticker = ['TECHM.NS', 'M&M.NS', 'WIPRO.NS', 'SBIN.NS', 'BAJFINANCE.NS',
    'INFY.NS', 'KOTAKBANK.NS', 'TRENT.NS', 'BAJAJFINSV.NS', 'SHRIRAMFIN.NS',
    'ADANIPORTS.NS', 'TATAMOTORS.NS', 'AXISBANK.NS', 'COALINDIA.NS', 'HCLTECH.NS',
    'BRITANNIA.NS', 'ICICIBANK.NS', 'LT.NS', 'CIPLA.NS', 'NTPC.NS',
    'BHARTIARTL.NS', 'NESTLEIND.NS', 'TATACONSUM.NS', 'HEROMOTOCO.NS', 'APOLLOHOSP.NS',
    'BEL.NS', 'HDFCBANK.NS', 'HINDALCO.NS', 'TCS.NS', 'SBILIFE.NS',
    'POWERGRID.NS', 'ULTRACEMCO.NS', 'ADANIENT.NS', 'HDFCLIFE.NS', 'GRASIM.NS',
    'ITC.NS', 'DRREDDY.NS', 'RELIANCE.NS', 'ONGC.NS', 'JSWSTEEL.NS',
    'EICHERMOT.NS', 'TITAN.NS', 'SUNPHARMA.NS', 'INDUSINDBK.NS', 'BPCL.NS',
    'MARUTI.NS', 'TATASTEEL.NS', 'BAJAJ-AUTO.NS', 'ASIANPAINT.NS', 'HINDUNILVR.NS']
```

```python
# ticker = [ 'HEROMOTOCO.NS','BAJAJ-AUTO.NS', 'TATASTEEL.NS']
```

```python
for company in ticker:
    data = yf.download(company, start = start, end = end)
    close_data = data['Close']

    #calculate the moving average to smooth data
    data['moving_avg'] = data['Close'].rolling(window=20).mean()
    #filling the blank spaces in the moving average usign the cumulative mean
    data['moving_avg'] = data['moving_avg'].fillna(data['Close'].expanding().mean())

  # Preprocess the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data['moving_avg'].values.reshape(-1, 1))

    # Create the training data
    #train_data = scaled_data[:-forecast_days]  # Use all but the last 20 days for training
    train_data = scaled_data
    X_train, y_train = [], []
    for i in range(past_date_to_use, len(train_data)):
        X_train.append(train_data[i-past_date_to_use:i, 0])
        y_train.append(train_data[i, 0])

    X_train, y_train = np.array(X_train), np.array(y_train)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

    # Create the LSTM model
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))  # Predicting the next price

    # Compile and train the model
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=100, batch_size=32)

    # Make predictions for the next 20 days
    inputs = scaled_data[len(scaled_data) - len(data) - past_date_to_use:].reshape(-1, 1)  # Prepare the input for the last 60 days
    inputs = np.concatenate((inputs, scaled_data[-forecast_days:]))  # Add the last 20 days to the inputs

    X_test = []
    for i in range(past_date_to_use, len(inputs)):
        X_test.append(inputs[i-past_date_to_use:i, 0])

    X_test = np.array(X_test)
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    predicted_prices = model.predict(X_test)
    predicted_prices = scaler.inverse_transform(predicted_prices)  # Reverse the scaling

    compare_data = yf.download(company, start = test_start, end = test_end)
    comp_data = compare_data['Close'].iloc[:forecast_days]
    predicted_dates = comp_data.index[:forecast_days]

    # Plotting the results
    plt.figure(figsize=(14, 5))
    plt.plot(data.index[-past_date_to_use:], data['Close'].values[-past_date_to_use:], color='blue', label='Historical Prices')
    #predicted_dates = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=forecast_days)
    #plt.plot(predicted_dates,comp_data, color = 'green', label = 'Actual Prices')
    plt.plot(predicted_dates, predicted_prices, color='red', label='Predicted Prices')
    plt.title(f'Stock Price Prediction: {company}')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.savefig(f'/content/drive/MyDrive/Ph.D/Datasets/LSTM/predictions/{company}.png', format='png', dpi=300)  # Adjust dpi for quality

    final_data = comp_data

    final_data = pd.DataFrame(final_data)

    final_data['Predicted'] = predicted_prices

    # final_data = final_data.rename(columns={'Predicted': 'Forecast'})

    # final_data['Returns'] = final_data['Forecast'].diff()

    # final_data.iloc[0, final_data.columns.get_loc('Returns')] = final_data.iloc[0,final_data.columns.get_loc('Close')]

    # final_data['Predicted'] = final_data['Returns'].cumsum()

    plt.figure(figsize=(14, 5))
    plt.plot(comp_data.index, comp_data, color='blue', label='Actual Prices')
    # predicted_dates = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=20)
    # predicted_dates = pd.date_range(start=comp_data.index[0]+pd.Timedelta(days=1), periods=20)
    # predicted_date = comp_data.index[:19]
    plt.plot(predicted_dates, final_data['Predicted'], color='red', label='Predicted Prices')
    plt.title(f'Prediction Comparison: {company}')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.savefig(f'/content/drive/MyDrive/Ph.D/Datasets/LSTM/comparison/{company}.png', format='png', dpi=300)  # Adjust dpi for quality
    plt.show()


    final_data.to_csv(f'/content/drive/MyDrive/Ph.D/Datasets/LSTM/comparison/{company}.csv')
```
